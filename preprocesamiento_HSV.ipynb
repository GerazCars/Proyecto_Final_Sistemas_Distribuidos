```python
import cv2
import os
```

```python
# Carpeta original
input_dir = "/content/drive/MyDrive/Classroom/SISTEMAS DISTRIBUIDOS ENE-JUN2026/Proyecto_Final/flowers"
```

```python
# Carpeta salida
output_dir = "/content/drive/MyDrive/Classroom/SISTEMAS DISTRIBUIDOS ENE-JUN2026/Proyecto_Final/flowers_preprocesadas_HSV"
```

```python
# Tamaño requerido por la CNN
IMG_SIZE = (150, 150)

# CREAR CARPETA DE SALIDA
os.makedirs(output_dir, exist_ok=True)

# RECORRER TODAS LAS CLASES
for clase in os.listdir(input_dir):
    ruta_entrada = os.path.join(input_dir, clase)
    if not os.path.isdir(ruta_entrada):
        continue
    ruta_salida = os.path.join(output_dir, clase)
    os.makedirs(ruta_salida, exist_ok=True)
    print(f"Procesando {clase}...")

# RECORRER IMÁGENES
    for archivo in os.listdir(ruta_entrada):
        ruta_imagen = os.path.join(ruta_entrada,archivo)
        try:
            img = cv2.imread(ruta_imagen)
            if img is None:
                continue

# REDIMENSIONAR
            img = cv2.resize(img,IMG_SIZE)

# CONVERSIÓN A HSV
            hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

# ECUALIZACIÓN DEL CANAL V
            hsv[:, :, 2] = cv2.equalizeHist(hsv[:, :, 2])

# REGRESAR A BGR
            img_resultado = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)

# GUARDAR
            salida = os.path.join(ruta_salida,archivo)
            cv2.imwrite(salida,img_resultado)
        except Exception as e:
            print(f"Error en {archivo}: {e}")

print("\nProceso terminado correctamente.")
print(f"Dataset guardado en: {output_dir}")
```
